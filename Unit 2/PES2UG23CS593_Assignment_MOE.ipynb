# Unit 2 Assignment: Building a Mixture of Experts (MoE) Router (Groq API)

```python
%pip install langchain_groq python-dotenv

from dotenv import load_dotenv
load_dotenv()

import getpass
import os
from langchain_groq import ChatGroq

if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API Key: ")
```

```python
MODEL_CONFIG = {
    "technical": {
        "system_prompt": "You are a technical support expert. Be precise, explain errors clearly, and give code fixes.",
        "temperature": 0.7
    },
    "billing": {
        "system_prompt": "You are a billing support expert. Be empathetic and explain refund or subscription policies clearly.",
        "temperature": 0.7
    },
    "general": {
        "system_prompt": "You are a helpful general assistant.",
        "temperature": 0.7
    }
}
```

```python
def route_prompt(user_input):

    router_llm = ChatGroq(
        model="llama-3.1-8b-instant",
        temperature=0
    )

    router_prompt = f"""
    Classify this text into one of these categories:
    technical, billing, general, tool

    Return ONLY the category name.

    Text: {user_input}
    """

    result = router_llm.invoke(router_prompt).content.strip().lower()

    if result not in ["technical", "billing", "general", "tool"]:
        return "general"

    return result
```

```python
def fetch_bitcoin_price():
    return "Current Bitcoin price (mock): $65,000"
```

```python
def process_request(user_input):

    category = route_prompt(user_input)

    if category == "tool":
        return fetch_bitcoin_price()

    config = MODEL_CONFIG[category]

    expert_llm = ChatGroq(
        model="llama-3.1-8b-instant",
        temperature=config["temperature"]
    )

    final_prompt = f"""
    {config['system_prompt']}

    User: {user_input}
    """

    response = expert_llm.invoke(final_prompt)

    return response.content
```

```python
print(process_request("My python script gives IndexError"))
print("-----------------------------------")
print(process_request("I was charged twice this month"))
print("-----------------------------------")
print(process_request("What is the current price of Bitcoin?"))
```
